```python
from glob import glob
import os
import pandas as pd

train_directory= '/content/drive/MyDrive/thesis/dataset/paddy_doctor/train_images'
train_files= list(glob(os.path.join(train_directory, '*', '*.jpg')))
train_labels= list(map(lambda x: x.split('/')[-2], train_files))
original_df= pd.DataFrame(list(zip(train_files, train_labels)), columns= ['Path','Label'])
original_df.head()
```

```python
# Plotting using countplot
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.figure(figsize= (10,5))
plt.title('No. observations for each class')
plt.xticks(rotation= 30, ha= 'right')
sns.countplot(data= original_df, x= 'Label',  color="tab:blue")
plt.show()
```

```python
labels= list(original_df.Label.unique())
labels
```

```python
from basic_image_eda import BasicImageEDA


def image_eda(path):
  if __name__ == '__main__':
      data_dir= path
      extensions= ['jpg']
      threads= 0
      dimension_plot= True
      channel_hist= True
      nonzero= False
      hw_division_factor= 1.0
      BasicImageEDA.explore(data_dir, extensions, threads, dimension_plot, channel_hist, nonzero, hw_division_factor)
```

```python
for label in labels:
  print('########## Analyzing for class: {} ##########'.format(label))
  path=  '/content/drive/MyDrive/thesis/dataset/paddy_doctor/train_images'+'/'+label
  print('reading images from path: {}'.format(path))
  image_eda(path)
```
